```python
import pandas as pd
import numpy as np

datasets = [3,4]
POP_SIZE = 20

df = pd.read_csv("Results/uniform/50x1000unit/100sensors/First_vs_Last_solutions.csv")
sets = len(df) / len(datasets) / POP_SIZE
sets = int(sets)

f1 = np.zeros((POP_SIZE,sets))
f2 = np.zeros((POP_SIZE,sets))
f3 = np.zeros((POP_SIZE,sets))

f1_last = np.zeros((POP_SIZE,sets))
f2_last = np.zeros((POP_SIZE,sets))
f3_last = np.zeros((POP_SIZE,sets))

dataset_id = 1
for j in range(sets):
    for i in range(POP_SIZE):
        f1[i][j]= df['Power_first'][i+(j+dataset_id)*POP_SIZE]
        f2[i][j]= df['Num. Active Sensors_first'][i+(j+dataset_id)*POP_SIZE]
        f3[i][j]= df['Avg. Dist. to Sink node_first'][i+(j+dataset_id)*POP_SIZE]
        f1_last[i][j]= df['Power_last'][i+(j+dataset_id)*POP_SIZE]
        f2_last[i][j]= df['Num. Active Sensors_last'][i+(j+dataset_id)*POP_SIZE]
        f3_last[i][j]= df['Avg. Dist. to Sink node_last'][i+(j+dataset_id)*POP_SIZE]

mean_f1 = np.mean(f1, axis=1)
mean_f2 = np.mean(f2, axis=1)
mean_f3 = np.mean(f3, axis=1)

std_f1 = np.std(f1, axis=1)
std_f2 = np.std(f2, axis=1)
std_f3 = np.std(f3, axis=1)

mean_f1_last = np.mean(f1_last, axis=1)
mean_f2_last = np.mean(f2_last, axis=1)
mean_f3_last = np.mean(f3_last, axis=1)

std_f1_last = np.std(f1_last, axis=1)
std_f2_last = np.std(f2_last, axis=1)
std_f3_last = np.std(f3_last, axis=1)

# print(f1)

# merge the results to df
df_out = pd.DataFrame()
df_out['Power_first_mean'] = mean_f1
df_out['Power_first_std'] = std_f1
df_out['Num. Active Sensors_first_mean'] = mean_f2
df_out['Num. Active Sensors_first_std'] = std_f2
df_out['Avg. Dist. to Sink node_first_mean'] = mean_f3
df_out['Avg. Dist. to Sink node_first_std'] = std_f3
df_out['Power_last_mean'] = mean_f1_last
df_out['Power_last_std'] = std_f1_last
df_out['Num. Active Sensors_last_mean'] = mean_f2_last
df_out['Num. Active Sensors_last_std'] = std_f2_last
df_out['Avg. Dist. to Sink node_last_mean'] = mean_f3_last
df_out['Avg. Dist. to Sink node_last_std'] = std_f3_last
df_out.to_csv(f"Results/uniform/50x1000unit/100sensors/First_vs_last_solutions_mean_std_{datasets[dataset_id]}.csv")
```

```python
# plot the result
import matplotlib.pyplot as plt
plt.errorbar(range(POP_SIZE), mean_f1, yerr=std_f1, fmt='o', label='Power_first')
plt.errorbar(range(POP_SIZE), mean_f1_last, yerr=std_f1_last, fmt='o', label='Power_last')
plt.xticks(range(POP_SIZE))  
plt.legend()
plt.show()

plt.errorbar(range(POP_SIZE), mean_f2, yerr=std_f2, fmt='o', label='Num. Active Sensors_first')
plt.errorbar(range(POP_SIZE), mean_f2_last, yerr=std_f2_last, fmt='o', label='Num. Active Sensors_last')
plt.xticks(range(POP_SIZE))  
plt.legend()
plt.show()

plt.errorbar(range(POP_SIZE), mean_f3, yerr=std_f3, fmt='o', label='Avg. Dist. to Sink node_first')
plt.errorbar(range(POP_SIZE), mean_f3_last, yerr=std_f3_last, fmt='o', label='Avg. Dist. to Sink node_last')
plt.xticks(range(POP_SIZE))  
plt.legend()
plt.show()
```
